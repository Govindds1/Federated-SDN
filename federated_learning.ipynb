```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression

def preprocess_and_train(df):
    X = df.drop(columns=['Label'])
    y = df['Label']

    le = LabelEncoder()
    y_encoded = le.fit_transform(y)

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y_encoded, test_size=0.2, random_state=42
    )

    model = LogisticRegression(max_iter=1000)
    model.fit(X_train, y_train)

    return model, X_test, y_test, le
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
import pandas as pd

# Load DDoS data
ddos_df = pd.read_csv('./Output/ddos_data.csv')

# Split features and labels
X_ddos = ddos_df.drop('Label', axis=1)
y_ddos = ddos_df['Label']

# Encode labels
le_ddos = LabelEncoder()
y_ddos_encoded = le_ddos.fit_transform(y_ddos)

# ðŸš€ Scale features to improve convergence
scaler = StandardScaler()
X_ddos_scaled = scaler.fit_transform(X_ddos)

# Train-test split
X_ddos_train, X_ddos_test, y_ddos_train, y_ddos_test = train_test_split(
    X_ddos_scaled, y_ddos_encoded, test_size=0.2, random_state=42
)

# Train model with higher max_iter
model_ddos = LogisticRegression(max_iter=3000)
model_ddos.fit(X_ddos_train, y_ddos_train)
```

```python
from sklearn.metrics import classification_report

y_pred_ddos = model_ddos.predict(X_ddos_test)
print(classification_report(y_ddos_test, y_pred_ddos))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Load Probe data
probe_df = pd.read_csv('./Output/probe_data.csv')

# Split features and labels
X_probe = probe_df.drop('Label', axis=1)
y_probe = probe_df['Label']

# Encode labels
le_probe = LabelEncoder()
y_probe_encoded = le_probe.fit_transform(y_probe)

# Scale features
scaler = StandardScaler()
X_probe_scaled = scaler.fit_transform(X_probe)

# Train-test split
X_probe_train, X_probe_test, y_probe_train, y_probe_test = train_test_split(
    X_probe_scaled, y_probe_encoded, test_size=0.2, random_state=42
)

# Initialize and train Logistic Regression model with increased max_iter
model_probe = LogisticRegression(max_iter=5000)
model_probe.fit(X_probe_train, y_probe_train)

# Predict on test data
y_pred_probe = model_probe.predict(X_probe_test)

# Print classification report
print("Classification Report for Probe Data:")
print(classification_report(y_probe_test, y_pred_probe))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import pandas as pd

# Load DoS data
dos_df = pd.read_csv('./Output/dos_data.csv')

# Split features and labels
X_dos = dos_df.drop('Label', axis=1)
y_dos = dos_df['Label']

# Encode labels
le_dos = LabelEncoder()
y_dos_encoded = le_dos.fit_transform(y_dos)

# Feature scaling
scaler = StandardScaler()
X_dos_scaled = scaler.fit_transform(X_dos)

# Train-test split
X_dos_train, X_dos_test, y_dos_train, y_dos_test = train_test_split(
    X_dos_scaled, y_dos_encoded, test_size=0.2, random_state=42
)

# Train Logistic Regression model with increased iterations
model_dos = LogisticRegression(max_iter=3000)
model_dos.fit(X_dos_train, y_dos_train)

# Predictions and evaluation
y_pred_dos = model_dos.predict(X_dos_test)
print("Classification Report for DoS Data:")
print(classification_report(y_dos_test, y_pred_dos))
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# Extract coefficients and intercepts
coef_ddos = model_ddos.coef_
coef_probe = model_probe.coef_
coef_dos = model_dos.coef_

intercept_ddos = model_ddos.intercept_
intercept_probe = model_probe.intercept_
intercept_dos = model_dos.intercept_

# Average the weights (coefficients) and intercepts
avg_coef = np.mean([coef_ddos, coef_probe, coef_dos], axis=0)
avg_intercept = np.mean([intercept_ddos, intercept_probe, intercept_dos], axis=0)

# Create new global model with averaged parameters
global_model = LogisticRegression(max_iter=3000)
global_model.coef_ = avg_coef
global_model.intercept_ = avg_intercept
global_model.classes_ = model_ddos.classes_  # assuming same label encoding

print("âœ… Federated Averaging completed. Global model is ready.")
```

```python
from sklearn.metrics import classification_report

y_pred_global = global_model.predict(X_ddos_test)
print("Classification Report for Global Model on DDoS Test Data:")
print(classification_report(y_ddos_test, y_pred_global))
```

```python
# DDoS test
print("DDoS Test:")
print(classification_report(y_ddos_test, global_model.predict(X_ddos_test)))

# Probe test
print("Probe Test:")
print(classification_report(y_probe_test, global_model.predict(X_probe_test)))

# DoS test
print("DoS Test:")
print(classification_report(y_dos_test, global_model.predict(X_dos_test)))
```

```python
import pandas as pd

# Load the datasets
ddos_df = pd.read_csv('output/ddos_data.csv')
probe_df = pd.read_csv('output/probe_data.csv')
dos_df = pd.read_csv('output/dos_data.csv')

# Update labels by grouping minor classes into 'Other'
# For DDoS: 'BFA', 'Normal' â†’ 'Other'; Keep 'DDoS'
ddos_map = {
    'DDoS': 'DDoS',
    'BFA': 'Other',
    'Normal': 'Other'
}
ddos_df['Label'] = ddos_df['Label'].map(ddos_map)

# For Probe: 'Probe' â†’ 'Probe'; 'Normal', 'Web-Attack' â†’ 'Other'
probe_map = {
    'Probe': 'Probe',
    'Web-Attack': 'Other',
    'Normal': 'Other'
}
probe_df['Label'] = probe_df['Label'].map(probe_map)

# For DoS: 'DoS' â†’ 'DoS'; 'BOTNET', 'Normal' â†’ 'Other'
dos_map = {
    'DoS': 'DoS',
    'BOTNET': 'Other',
    'Normal': 'Other'
}
dos_df['Label'] = dos_df['Label'].map(dos_map)

# Optionally save the updated versions
ddos_df.to_csv('output/ddos_relabelled.csv', index=False)
probe_df.to_csv('output/probe_relabelled.csv', index=False)
dos_df.to_csv('output/dos_relabelled.csv', index=False)

print("âœ… Relabeling done and saved.")
```
